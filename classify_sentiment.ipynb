##### Import libs

```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

##### Read opinions from the csv file

```python
opinions = pd.read_csv('./consolidated_opinions.csv', encoding='UTF-8')
```

```python
opinions = opinions[['opinion','score','sentiment']]
```

```python
opinions.sentiment = opinions.sentiment.str.lower()
```

```python
sa = SentimentIntensityAnalyzer()
# opinions.vader = SentimentIntensityAnalyzer()
opinions['vader'] = opinions.opinion.apply(lambda o: sa.polarity_scores(o))
```
